```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('/kaggle/input/house-prices-advanced-regression-techniques/train.csv')
```

```python
data.head(10)
```

```python
data.tail(10)
```

```python
# Calculate the correlation matrix
correlation = data[['LotArea', 'YrSold', 'SalePrice']].corr()

# Visualize the correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(correlation, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.violinplot(x='MSZoning', y='SalePrice', data=data)
plt.xlabel('MSZoning')
plt.ylabel('SalePrice')
plt.title('Comparison of MSZoning by SalePrice')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.lineplot(x='YrSold', y='SalePrice', data=data)
plt.xlabel('YrSold')
plt.ylabel('SalePrice')
plt.title('Variation of SalePrice During the Years')
plt.show()
```
